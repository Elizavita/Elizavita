```python
# Блок 2
# Задание 1

# Второй параметр "human"

series = dataSet['nature']
series2 = dataSet['human']
seriesHistogram = np.histogram(series)

seriesInter=[]
series2Gr=[]
for i in range(len(seriesHistogram[1])-1):
    group=[]
    j=0
    left=seriesHistogram[1][i]
    right = seriesHistogram[1][i+1]
    seriesInter.append(str(left)+'-'+str(right))
    for val in series:
        if left<val<=right:
            group.append(series2[j])
            j+=1
    series2Gr.append(np.array(group))
    
series2Groups=pd.DataFrame()
series2Groups['интервалы']=pd.Series(seriesInter)
series2Groups['частоты']=pd.Series([len(x)  for x in series2Gr])
series2Groups['среднее']=pd.Series([ (x.mean() if len(x)>0 else 0) for x in series2Gr])
series2Groups['дисперсия']=pd.Series([ (x.std()**2 if len(x)>0 else 0) for x in series2Gr])

display(series2Groups)
```

```python
# Задание 2

D_in_avg = (series2Groups['дисперсия']*series2Groups['частоты']).sum()/series2Groups['частоты'].sum()
D_out = (((series2Groups['среднее']-dataSet['human'].mean())**2)*series2Groups['частоты']).sum()/series2Groups['частоты'].sum()
print('средняя внутригрупповая дисперсия + межгрупповая дисперсия ',D_in_avg+D_out)
print('дисперсия human ',dataSet['human'].std()**2)

EmpCorr = sqrt(1-D_in_avg/dataSet['human'].std()**2)
print('эмпирическое корреляционное соотношение ',EmpCorr)
```

```python
# Задание 4

series = dataSet['nature']
series2 = dataSet['human']
PirsonCorr = ((series*series2).mean()-series.mean()*series2.mean())/(series.std()*series2.std())
print('коэффициент корреляции Пирсона = ', PirsonCorr)

from scipy.stats import t
Alpha = 0.05
DegreeFreedom = len(series)-2                               
myT = PirsonCorr/(sqrt(1-PirsonCorr**2))*sqrt(series.shape[0]-2)
critT = t.ppf(1-Alpha/2,4)
print('наблюдаемое значение T ',myT)
print('критическое значение Т ',critT)
```

```python
# Задание 5

dataSet_numbers = dataSet.select_dtypes(['float','int']).copy()
display(dataSet_numbers.head())

corr = np.array(dataSet_numbers.corr())
def matrix_minor(arr, i, j):
    return np.delete(np.delete(arr,i,axis=0), j, axis=1)

columns_NameIndex = {k:v for k,v in zip(dataSet_numbers.columns,list(range(dataSet_numbers.shape[1])))}

i=columns_NameIndex['nature']
j=columns_NameIndex['human']
A_ij = (-1)**(i+j)*np.linalg.det(matrix_minor(corr,i,j))
A_ii = np.linalg.det(matrix_minor(corr,i,i))
A_jj = np.linalg.det(matrix_minor(corr,j,j))
partCorr = -A_ij/sqrt(A_ii*A_jj)
print('частный коэффициент корреляции = ',partCorr)
```

```python
# Задание 6

#Ранговый коэффициент корреляции Спирмена

from collections import Counter
df_rank = dataSet[['nature','human']].rank()
def Tx(series):
    freq_list = np.array(list(Counter(series).values()))
    T = 0.5*(freq_list**3-freq_list).sum()
    return T

a = 1/6*(df_rank.shape[0]**3 - df_rank.shape[0])
series = ((df_rank['nature']-df_rank['human'])**2).sum()
Tx1 = Tx(dataSet['nature'])
Tx2 = Tx(dataSet['human'])
SpearmanCorr = (a-series-Tx1-Tx2) / (sqrt((a- 2*Tx1)* (a - 2*Tx2)))
T_crit = t.ppf(1 - 0.05, df_rank.shape[0]-2)*sqrt((1-SpearmanCorr**2)/(df_rank.shape[0]-2))

print('наблюдаемное зачение коэфф. Спирмена ',abs(SpearmanCorr))
print('критическое значение ',T_crit)
```

Наблюдаемое значение больше критического. связи между параметрами "nature" и "human" нет.

```python
#Ранговый коэффициент корреляции Кендалла
def MyKendall(dataFrame,columns):
    P=0
    Q=0

    df_rank = dataFrame[columns].rank()
    df_rank_sort = df_rank.sort_values(by=columns[0],axis=0, ascending=True)
    
    for i in range(df_rank_sort.shape[0]):
        df_1 = df_rank_sort.iloc[i+1:,1]
        P+= df_1[df_1 > df_rank_sort.iloc[i,1]].count()
    
    for i in range(df_rank_sort.shape[0]):
        df_1 = df_rank_sort.iloc[i+1:,1]
        Q+= df_1[df_1 < df_rank_sort.iloc[i,1]].count()
    
    S=P-Q
    n = df_rank_sort.shape[0]
    
    return 2*S/(n*(n-1))

KebndallCorr = MyKendall(dataSet,['nature','human'])
print('коэфф. корреляции Кендалла = ',KebndallCorr)
```
